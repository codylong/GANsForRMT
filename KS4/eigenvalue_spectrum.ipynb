# Eigenvalue Spectrum

We endeavor to understand better the eigenvalue spectrum of KS Kahler metrics, so that we might be able to learn a random matrix approximation of them that scales well with h11.

## Data import

We check to make sure that the number of eigenvalues is h11, a non-trivial bug check

```python
import os
import cPickle as pickle
import numpy as np

files = [d for d in os.listdir('data') if d[0] == 'h' and 'evals' in d]
print('num files', len(files))
assert_mismatch = []
def info_from_file(file):
    h11 = int(file.split("_")[1])
    s = file.split("_")[2]
    poly = int(
    s[s.index('y')+1:])
    f = open("data/"+file,'rb')
    evals = pickle.load(f)
    f.close()
    to_ret = [(h11,poly),evals]
    try:
        assert len(evals) == h11
    except AssertionError:
        assert_mismatch.append(to_ret[0])
    return to_ret

raw_data = [info_from_file(file) for file in files]
print(set([len(r[1]) - r[0][0] for r in raw_data]))
raw_data = [r for r in raw_data if len(r[1]) - r[0][0] == 0]
print("num files after eval mismatch subtraction", len(raw_data))
print(assert_mismatch)
```

```python
#first let's plot the various spectra on top of one another, in order to visualize the N-behavior of the bump
def getevals(h11):
    evs = []
    for r in raw_data:
        if r[0][0] == h11:
            evs = evs + list([np.log10(ii) for ii in r[1]])
    return evs
e10 = getevals(10)
e20 = getevals(20)
e30 = getevals(30)
e40 = getevals(40)
e50 = getevals(50)
```

```python
#what we're seeing is 1) the size of the bump grows with N, 2) 
#the position of the bump shifts with N, 3) the position of main peak shifts with N
#reasonble guess: this has something to do with the stretched kahler cone

#let's look at the smallest eigenvalues first.
def getmins(h11):
    maxes = []
    for r in raw_data:
        if r[0][0] == h11:
            maxes = maxes + [np.log10(min(r[1]))]
    return maxes
m10 = getmins(10)
m20 = getmins(20)
m30 = getmins(30)
m40 = getmins(40)
m50 = getmins(50)
```

```python
from scipy.signal import find_peaks
#print len(sortyspre)

#need to find actual peaks. also ensure that they're not in the tail. here's the first function that does so
#(we'll use two)
def get_sharp_peaks1(data):
    f = sns.kdeplot(data)
    x,y = f.get_lines()[0].get_data()
    plt.clf()
    #print y
    ny=[-k for k in y]
    peak_idxs, trough_idxs = find_peaks(y)[0], find_peaks(ny)[0]
    peak_locs, trough_locs = [x[k] for k in peak_idxs], [x[k] for k in trough_idxs]
    npeaks = len(peak_idxs)
    idxs = list(peak_idxs) + list(trough_idxs)
    idxs.sort()
    actual_peaks = []
    #first we check that they're peaky enough
    for idx in idxs:
        if idx in peak_idxs:
            peak_idx, peak_x, peak_y = list(peak_idxs).index(idx), x[idx], y[idx]
            #print 'peak_x', peak_x
            if len(trough_idxs) > 0 and peak_idx < len(trough_idxs):
                #print peak_x, peak_idx, trough_idxs
                trough_y = y[trough_idxs[peak_idx]] # next trough at same idx
                #print peak_y, trough_y,  abs(trough_y/peak_y)
                assert peak_y > trough_y
                if abs(trough_y/peak_y) < .5:
                    #print 'RETURNING TRUE'
                    actual_peaks.append(x[idx])
            else: actual_peaks.append(x[idx])
    #now we make sure they're not in the tail:
    
    good_peaks = []
    for pp in actual_peaks:
        if pp > -2:
            good_peaks.append(pp)
    return good_peaks

def get_sharp_peaks2(data,thresh = 1.5):
    data_list = list(data)
    f = sns.kdeplot(data)
    x,y = f.get_lines()[0].get_data()
    plt.clf()
    #print y
    ny=[-k for k in y]
    peak_idxs, trough_idxs = find_peaks(y)[0], find_peaks(ny)[0]
    peak_locs, trough_locs = [x[k] for k in peak_idxs], [x[k] for k in trough_idxs]
    npeaks = len(peak_idxs)
    idxs = list(peak_idxs) + list(trough_idxs)
    idxs.sort()
    actual_peaks = []
    #first we check that they're peaky enough
    for idx in idxs:
        if idx in peak_idxs:
            peak_idx, peak_x, peak_y = list(peak_idxs).index(idx), x[idx], y[idx]
            if sum([1 if (e <= peak_x + thresh and e >= peak_x - thresh) else 0 for e in data_list]) > 6:
                actual_peaks.append(peak_x)
    #now we make sure they're not in the tail:
    good_peaks = []
    for pp in actual_peaks:
        if pp > -2:
            good_peaks.append(pp)
    return good_peaks
#test them both:
print get_sharp_peaks1(sortyspre50[1])
print get_sharp_peaks2(sortyspre50[1])
sns.distplot(sortyspre50[1],
             kde=True, label='0', norm_hist = True)
plt.legend(prop={'size':12})
plt.show()
```
